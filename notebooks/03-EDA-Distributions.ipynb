### EDA

```python
import pandas as pd
import numpy as np
import os

data_path = os.path.join(os.path.pardir, 'data/raw/')
train_path = data_path + 'train.csv'
test_path = data_path + 'test.csv'
train_df = pd.read_csv(train_path, index_col='PassengerId')
test_df = pd.read_csv(test_path, index_col='PassengerId')

test_df['Survived'] = -1
df = pd.concat((train_df, test_df))
df = train_df
```

## Distributions

```python
train_df.Age.plot(kind='hist', color='c', bins=20)
```

```python
train_df.Age.plot(kind='kde', color='c')
```

```python
print(df.Age.skew())
print(df.Fare.skew())
```

```python
df.plot.scatter(x='Pclass', y='Fare', alpha=0.1)
```

```python
df[df.Pclass == 3].Age.max()
```

```python
df[df.Survived == 0].Pclass.plot(kind='hist', bins=3)
```

```python
df.Pclass.plot(kind='hist', bins=3)
```

# Gropping

```python
df.groupby('Pclass').agg({'Age': 'median', 'Fare': 'median'})
```

```python
df.groupby(['Pclass', 'Survived', 'Sex']).agg({'Age': 'median', 'Fare': 'median', 'Sex': 'count'})
```

# CrossTab

```python
pd.crosstab(df.Survived, df.Pclass)
```

```python
pd.crosstab(df.Survived, df.Pclass).plot(kind='bar')
```

# Pivot

```python
df.pivot_table(index='Sex', columns='Pclass', values='Age', aggfunc='mean')
```

```python
df.groupby(['Pclass', 'Sex']).Age.mean().unstack()
```
